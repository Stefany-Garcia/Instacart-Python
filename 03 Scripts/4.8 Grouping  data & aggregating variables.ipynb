## Question 4

```python
#use the transform() and loc() functions to create a loyalty flag for customers with many orders 
```

```python
#creating a column for the max number of orders per customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
#create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

## Question 5

```python
# Check basis statistics of product prices for each loyalty category
ords_prods_merge.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

## Question 6

```python
# create a column of mean price per customer
ords_prods_merge['mean_order'] = ords_prods_merge.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head(15)
```

```python
ords_prods_merge['mean_order'].describe()
```

```python
# create the flag that targets low and high spenders
ords_prods_merge.loc[ords_prods_merge['mean_order'] < 10, 'spending_flag'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['mean_order'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
ords_prods_merge[['customer_id', 'mean_order', 'spending_flag']].head(15)
```

## Question 7

```python
#create column for median days since prior order
ords_prods_merge['median_since_prior_order'] = ords_prods_merge.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge.head(15)
```

```python
# create flag for frequent customers
ords_prods_merge.loc[ords_prods_merge['median_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_since_prior_order'] > 10) & (ords_prods_merge['median_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
